```python
data = pd.read_csv(r"D:\Intellipaat\kegal datasets\steel plates dataset\train.csv")
data
```

```python
col_name = []
with open(r"D:\Intellipaat\kegal datasets\steel plates dataset\Faults27x7_var") as f:
    for line in f:
        col_name.append(line.strip())
```

```python
data  = data.drop(columns=['id'])
```

```python
org_data = pd.read_csv("D:\Intellipaat\kegal datasets\steel plates dataset\Faults.NNA",engine= "python",delimiter= '\s',names= col_name)
org_data
```

```python
df = pd.concat([data,org_data],ignore_index= True)
df
```

```python
x = df.iloc[:,:-7]
x
```

```python
y_target= pd.DataFrame({
    'Pastry': df['Pastry'],
    'Z_Scratch': df['Z_Scratch'],
    'K_Scatch': df['K_Scatch'],
    'Stains': df['Stains'],
    'Dirtiness': df['Dirtiness'],
    'Bumps': df['Bumps'],
    'Other_Faults': df['Other_Faults']
})
```

```python
std = StandardScaler()
X = std.fit_transform(x)
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y_target,test_size=0.2,random_state = 42)
```
