```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import RidgeClassifier
```

```python
df = pd.read_csv(r"D:\Intellipaat\kegal datasets\steel plates dataset\train.csv")
df
```

```python
x = df.iloc[:,:-7]
y= df.iloc[:,-7:]
```

```python
std = StandardScaler()
x_train = std.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_train,y,test_size=0.2,random_state=42)
```

```python
for i in range(7):
    Ridge = RidgeClassifier()
```

```python
from sklearn.multioutput import MultiOutputClassifier
moc = MultiOutputClassifier(Ridge)
moc
```

```python
moc.fit(x_train,y_train)
```

```python
y_pred = moc.predict(x_test)
y_pred
```

```python
roc_auc_score(y_test,y_pred)
```

```python
test_org= pd.read_csv(r"D:\Intellipaat\kegal datasets\steel plates dataset\test.csv")
test_org
```

```python
test = std.transform(test_org)
```

```python
predictions = pd.DataFrame(moc.predict(test))
predictions
```
