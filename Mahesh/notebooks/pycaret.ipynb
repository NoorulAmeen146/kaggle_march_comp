```python
from pycaret.classification import *

# Load data
data = pd.read_csv(r"D:\Intellipaat\kegal datasets\steel plates dataset\train.csv")

# Flatten the multi-label target columns into a single target column containing a list of labels
data['target'] = data[['Pastry','Z_Scratch','K_Scatch','Stains','Dirtiness','Bumps','Other_Faults']].apply(lambda x: ','.join(x.astype(str)), axis=1)

# Drop the original target columns
data.drop(['Pastry','Z_Scratch','K_Scatch','Stains','Dirtiness','Bumps','Other_Faults'], axis=1, inplace=True)

# Setup PyCaret environment with the single target column representing all labels
setup(data, target='target')

# Compare models
best_model = compare_models()
```

```python
from pycaret.classification import *
import pandas as pd

# Load data
data = pd.read_csv(r"D:\Intellipaat\kegal datasets\steel plates dataset\train.csv")

# Loop through each target column
for target_column in ['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults']:
    print(f"Processing target column: {target_column}")
    
    # Set up data for the current target column
    setup(data, target=target_column)

    # Compare models
    best_model = compare_models()
    
    # Print the best model for the current target column
    print(f"Best model for {target_column}: {best_model}")
```
